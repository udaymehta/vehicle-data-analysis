```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid", {'grid.linestyle': '-.'})
```

```python
df = pd.read_csv('cars.csv')
```

```python
df.head(10)
```

```python
df.describe()
```

```python
df.info()
```

```python
headers = ['mpg', 'cylinders', 'cubicinches', 'hp', 'weightlbs', 'time-to-60', 'year', 'brand']
df.columns = headers
df.columns
```

# 1.Can you find that there is improvement in mpg over the years ?


```python
sns.lineplot(x='year', y='mpg', data=df, hue="brand", style="brand", markers=True, dashes=True, errorbar=None)
```

```python
df['cubicinches'] = pd.to_numeric(df['cubicinches'], errors='coerce')
```

# 2.Does cubicinches of a vehicle affects the vehicles time to reach 60 Miles per hour?



```python
correlation = df['cubicinches'].corr(df['time-to-60'])
print(f"Correlation: {round(correlation, 3)}")
print('\n')

plt.figure(figsize=(10, 6))
sns.scatterplot(x='cubicinches', y='time-to-60', data=df)
```

# 3.Does a specific country prefer vehicles with more horse power?

```python
sns.barplot(x='brand', y='hp', data=df, errorbar=None)
```

# 4.How does the horse power affects the average of the vehicle ?

```python
sns.lineplot(x='hp', y='mpg', data=df, hue="brand", style="brand", markers=True, dashes=True, errorbar=None)
```

# 5.How does the cubic inches affects the average of the vehicle ?


```python
sns.lineplot(x='cubicinches', y='mpg', data=df, hue="brand", style="brand", markers=True, dashes=True, errorbar=None)
```

# 6.What is the effect of number of cylinders on the mpg?


```python
sns.boxplot(x='cylinders', y='mpg', hue='brand', data=df)
```

# 7.What relation between cubic inches , number of cylinders and horse power you can observe from the data?

```python
sns.lineplot(x='cubicinches', y='hp', data=df, hue="cylinders", style="cylinders", palette="flare", markers=True, dashes=False, errorbar=None)
```

```python
sns.pairplot(data=df, vars=['cubicinches', 'cylinders', 'hp'], hue='brand')
```

# 8.How much proportion of the total records is occupied by vehicles with 8 cylinders?

```python
eight_cylinder_vehicles = len(df[df['cylinders'] == 8])
total_vehicles = len(df)
proportion = eight_cylinder_vehicles / total_vehicles

print(f"Proportion of vehicles with 8 cylinders: {proportion:.2%}")
```
